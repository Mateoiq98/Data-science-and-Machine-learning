### EXTRACCION DE DATOS GENERALES IDST
 Se definen las librerias con las que trabajeron para el manejo de datos y la extracción.
 Se realiza un request para extraer los datos denotados en JSON.

```python
### EXTRACCION DE DATOS
import requests
import pandas as pd
import json
from seaborn import load_dataset
header = {
        "User-Agent": "some user-agent"}
resp = requests.post('https://idst.inl.gov/Home/Search/', headers=header)
data_electrolyte=resp.json()


```

### CONVERTIR DATOS A UN DATA FRAME PANDAS
Esto con la finalidad de manipular y tener mejor visualicion de los datos


```python
datos = pd.DataFrame(data_electrolyte)
datos.head()
```

### Ordenar ID  de los solutos
Para tener un mejor orden

```python
d_e = datos.sort_values('Solute_ID')
d_e.tail(20)
```

### Recolector de datos con el ID

```python
def insertarcolumnas(data_frame,new_columns,i):
    for etiqueta in reversed(new_columns):
        data_frame.insert(0,etiqueta,d_e.iloc[i][etiqueta])
list_etiqueta=["Solute_ID",'Formula','Solute_Mol_Mass','Num_Cations','Num_Anions','Cation','Anion','Num_Particles','Cation_Charge','Anion_Charge','Temperature_K','Solvent_Name']   
df=pd.DataFrame()
for i in range(len(d_e)):
    d_id=d_e.iloc[i]['Solute_ID']
    resp = requests.post('https://idst.inl.gov/Home/Data/'+str(d_id)+"/", headers=header)
    if str(resp) == "<Response [200]>":
        if d_id <= 950:
            data_iter=resp.json()
            data_iter=pd.DataFrame(data_iter)
            insertarcolumnas(data_iter,list_etiqueta,i)                                 
            df=pd.concat([df,data_iter])
df.head()
```
